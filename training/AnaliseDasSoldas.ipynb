```python
import math
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib widget
```

```python
imagemBoa = cv.imread("Soldas_boas/Solda_1.png")
imagemAusente = cv.imread("Soldas_ausente/Solda_1.png")
imagemExcesso = cv.imread("Soldas_excesso/Solda_1.png")
imagemPonte = cv.imread("Soldas_ponte/Solda_1.png")
imagemPouca = cv.imread("Soldas_pouca/Solda_1.png")
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3)
ax[0,0].imshow(imagemBoa)
ax[0,0].set_xticks([])
ax[0,0].set_yticks([])
ax[0,1].imshow(imagemAusente)
ax[0,1].set_xticks([])
ax[0,1].set_yticks([])
ax[0,2].imshow(imagemExcesso)
ax[0,2].set_xticks([])
ax[0,2].set_yticks([])
ax[1,0].imshow(imagemPonte)
ax[1,0].set_xticks([])
ax[1,0].set_yticks([])
ax[1,1].imshow(imagemPouca)
ax[1,1].set_xticks([])
ax[1,1].set_yticks([])
ax[1,2].imshow(imagemBoa)
ax[1,2].set_xticks([])
ax[1,2].set_yticks([])
```

```python
altura, largura = imagemBoa.shape[:2]
transla = np.float32([[1, 0, -3],[0, 1, -3]])
imagemBoaT = cv.warpAffine(imagemBoa, transla, (largura, altura),
                            flags= cv.INTER_CUBIC,
                            borderMode=cv.BORDER_WRAP)
imagemAusenteT = cv.warpAffine(imagemAusente, transla, (largura, altura),
                                flags= cv.INTER_CUBIC,
                                borderMode=cv.BORDER_WRAP)
imagemExcessoT = cv.warpAffine(imagemExcesso, transla, (largura, altura),
                                flags= cv.INTER_CUBIC,
                                borderMode=cv.BORDER_WRAP)
imagemPonteT = cv.warpAffine(imagemPonte, transla, (largura, altura),
                            flags= cv.INTER_CUBIC,
                            borderMode=cv.BORDER_WRAP)
imagemPoucaT = cv.warpAffine(imagemPouca, transla, (largura, altura),
                            flags= cv.INTER_CUBIC,
                            borderMode=cv.BORDER_WRAP)

fig, ax = plt.subplots(nrows=2, ncols=3)
ax[0,0].imshow(imagemBoaT)
ax[0,0].set_xticks([])
ax[0,0].set_yticks([])
ax[0,1].imshow(imagemAusenteT)
ax[0,1].set_xticks([])
ax[0,1].set_yticks([])
ax[0,2].imshow(imagemExcessoT)
ax[0,2].set_xticks([])
ax[0,2].set_yticks([])
ax[1,0].imshow(imagemPonteT)
ax[1,0].set_xticks([])
ax[1,0].set_yticks([])
ax[1,1].imshow(imagemPoucaT)
ax[1,1].set_xticks([])
ax[1,1].set_yticks([])
ax[1,2].imshow(imagemBoaT)
ax[1,2].set_xticks([])
ax[1,2].set_yticks([])
```
