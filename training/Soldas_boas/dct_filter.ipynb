```python
import math
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib widget
```

```python
imagem = cv.imread("Solda_1234.png")
fig = plt.figure()
ax = plt.axes()
ax.set_xticks([])
ax.set_yticks([])
ax.imshow(imagem)
```

```python
gray_img = cv.cvtColor(imagem, cv.COLOR_RGB2GRAY)
print(gray_img.shape)
print(gray_img.max())
print(gray_img.min())
```

```python
lut = np.asarray([math.log(i, 2) for i in (np.arange(0, 256) + 1)], "float64")
print(lut.shape)
print(lut.min())
print(lut.max())
```

```python
img_log = cv.LUT(gray_img, lut)
print(img_log.shape)
print(img_log.min())
print(img_log.max())
```

```python
dctImage = cv.dct(img_log)
print(dctImage.shape)
print(dctImage.min())
print(dctImage.max())
```
