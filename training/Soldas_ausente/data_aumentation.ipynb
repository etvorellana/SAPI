* Imagem Original

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib widget

imagem = cv.imread("Solda_1.png")
print (imagem.shape)
altura, largura = imagem.shape[:2]
ponto = (largura / 2, altura / 2) #ponto no centro da figura

fig = plt.figure()
ax = plt.axes()
ax.set_xticks([])
ax.set_yticks([])
ax.imshow(imagem)
```

* Primeira transformao: Rotao de 30 graus, testando os parmetros da rotao 

```python
#rotacao 30
rotacao = cv.getRotationMatrix2D(ponto, 30, 1.0)
rot30Img01 = cv.warpAffine(imagem, rotacao, (largura, altura), 
                        flags= cv.INTER_CUBIC, 
                        borderMode=cv.BORDER_CONSTANT)
rot30Img02 = cv.warpAffine(imagem, rotacao, (largura, altura),
                        flags= cv.INTER_CUBIC,
                        borderMode=cv.BORDER_REPLICATE)
rot30Img03 = cv.warpAffine(imagem, rotacao, (largura, altura), 
                        flags= cv.INTER_CUBIC,
                        borderMode=cv.BORDER_REFLECT)
rot30Img04 = cv.warpAffine(imagem, rotacao, (largura, altura),
                        flags= cv.INTER_CUBIC,
                        borderMode=cv.BORDER_WRAP )
rot30Img05 = cv.warpAffine(imagem, rotacao, (largura, altura),
                        flags= cv.INTER_CUBIC,
                        borderMode=cv.BORDER_REFLECT_101)
rot30Img06 = cv.warpAffine(imagem, rotacao, (largura, altura),
                        flags= cv.INTER_CUBIC,
                        borderMode=cv.BORDER_TRANSPARENT)
rot30Img07 = cv.warpAffine(imagem, rotacao, (largura, altura),
                        flags= cv.INTER_CUBIC,
                        borderMode=cv.BORDER_DEFAULT)

fig, ax = plt.subplots(nrows=2, ncols=4)
ax[0,0].imshow(imagem)
ax[0,0].set_xticks([])
ax[0,0].set_yticks([])
ax[0,1].imshow(rot30Img01)
ax[0,1].set_xticks([])
ax[0,1].set_yticks([])
ax[0,2].imshow(rot30Img02)
ax[0,2].set_xticks([])
ax[0,2].set_yticks([])
ax[0,3].imshow(rot30Img03)
ax[0,3].set_xticks([])
ax[0,3].set_yticks([])
ax[1,0].imshow(rot30Img04)
ax[1,0].set_xticks([])
ax[1,0].set_yticks([])
ax[1,1].imshow(rot30Img05)
ax[1,1].set_xticks([])
ax[1,1].set_yticks([])
ax[1,2].imshow(rot30Img06)
ax[1,2].set_xticks([])
ax[1,2].set_yticks([])
ax[1,3].imshow(rot30Img07)
ax[1,3].set_xticks([])
ax[1,3].set_yticks([])
```

```python
rot030Img = cv.warpAffine(imagem, rotacao, (largura, altura),
                        flags= cv.INTER_CUBIC,
                        borderMode=cv.BORDER_REPLICATE)
cv.imwrite("Solda_2.png", rot030Img)
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].imshow(imagem)
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[1].imshow(rot030Img)
ax[1].set_xticks([])
ax[1].set_yticks([])
```
